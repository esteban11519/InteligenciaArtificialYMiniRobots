### 1. Multi Expression Programming X 

Multi Expression Programming X (Multi Expression Programming X, s. f.) es un programa que permite general varias soluciones en un mismo cromosoma y usa una aproximación evolutiva. En esta parte se va a utilizar para predecir el precio del dolar. Los datos son de Yahoo Finance (USD/COP (COP=X) Live Rate, Chart & News - Yahoo Finance, s. f.) desde el 22 de noviembre de 2021 hasta el 21 de noviembre de 2022 y los datos de prueba son tomados desde el 21 de noviembre hasta el 6 de diciembre de 2022. Estas fueron las operaciones usadas en Multi Expression Programming X.

La función en python del modelo se presenta a continuación:

```python
# Librerias
import math
import numpy as np
import matplotlib.pyplot as plt
from gplearn.genetic import SymbolicRegressor
```

```python
# Funciones

def tablaDeEntrenamiento(precios,m):
    ''' 
    precios: Datos unidimencional
    m: Total de columnas incluento en target
    '''

    tabla=[]
    for i in range(len(precios)-m+1):
        tabla.append(precios[i:i+m])

    return tabla

def tablaDePrueba(precios,m):
    ''' 
    precios: Datos unidimencional
    m: Total de columnas incluento en target
    '''
    tabla=[]
    for i in range(len(precios)-m+1):
        tabla.append(precios[i:i+m-1])

    return tabla
```

```python
#Generated by libmep version 2022.10.1.0-beta
def mepx(x, outputs):
  prg = [0] * 50
  prg[0] = x[0]
  prg[1] = math.tan(prg[0])
  prg[2] = math.log(prg[0])
  prg[3] = x[2]
  prg[4] = prg[3] + prg[1]
  prg[5] = x[0]
  prg[6] = x[1]
  prg[7] = 1/prg[1]
  prg[8] = math.tan(prg[7])
  prg[9] = math.ceil(prg[4])
  prg[10] = x[2]
  prg[11] = math.pow(prg[5], prg[2])
  prg[12] = math.ceil(prg[8])
  prg[13] = max(prg[4], prg[8])
  prg[14] = prg[4] - prg[12]
  prg[15] = x[1]
  prg[16] = x[0]
  prg[17] = math.log2(prg[5])
  prg[18] = math.atan(prg[3])
  prg[19] = x[2]
  prg[20] = x[2]
  prg[21] = x[2]
  prg[22] = x[2]
  prg[23] = x[1]
  prg[24] = prg[18] if prg[17] < 0 else prg[14] # ifalzbc
  prg[25] = x[2]
  prg[26] = x[1]
  prg[27] = x[0]
  prg[28] = x[0]
  prg[29] = x[1]
  prg[30] = math.sin(prg[11])
  prg[31] = x[2]
  prg[32] = x[1]
  prg[33] = prg[24] + prg[30]
  prg[34] = max(prg[15], prg[26])
  prg[35] = math.ceil(prg[16])
  prg[36] = x[0]
  prg[37] = -prg[2]
  prg[38] = x[2]
  prg[39] = x[1]
  prg[40] = x[1]
  prg[41] = prg[32] if prg[11] < 0 else prg[11] # ifalzbc
  prg[42] = abs(prg[2])
  prg[43] = x[0]
  prg[44] = abs(prg[29])
  prg[45] = x[2]
  prg[46] = math.floor(prg[33])
  prg[47] = x[2]
  prg[48] = x[0]
  prg[49] = x[2]

  outputs[0] = prg[46]

#example of utilization ...

x = [
  3912.250000,
  3924.000000,
  3948.250000
]

outputs = [0]

mepx(x, outputs)
print(outputs[0])
```

### 2. Librería gplearn

gplearn es una extensión de scikit-learn usado para machine learn la cual realiza programación genética con regresión symbólica. (Examples — gplearn 0.4.2 documentation, s. f.; Introduction to GP — gplearn 0.4.2 documentation, s. f.; Stephens, 2015/2022)

#### Symbolic Regressor

```python
filepathcsv='./tablaDeEntrenamiento.csv'
data = np.genfromtxt(filepathcsv, delimiter='\t',skip_header=0)

D=np.arange(data.shape[0])+1
# Training samples
X_train = data[:,:3]
y_train = data[:,3]

filepathcsv='./tablaTestTotal.csv'
data = np.genfromtxt(filepathcsv, delimiter='\t',skip_header=0)


# Testing samples
X_test = data[:,:3]
y_test = data[:,3]
```

```python
est_gp = SymbolicRegressor(population_size=5000,
                           generations=20, stopping_criteria=0.01,
                           p_crossover=0.7, p_subtree_mutation=0.1,
                           p_hoist_mutation=0.05, p_point_mutation=0.1,
                           max_samples=0.9, verbose=1,
                           parsimony_coefficient=0.01, random_state=0)
est_gp.fit(X_train, y_train)
```

```python
print(est_gp._program)
y_trainPrediction=est_gp.predict(X_train)
y_testPrediction = est_gp.predict(X_test) 
```

```python
errorAbsoluto=y_train-y_trainPrediction
# Graficación

markerSize=3

fig,ax=plt.subplots()
ax.plot(D,y_train, 'ro', label='Real',markersize=markerSize)
ax.plot(D,y_trainPrediction,'go',label='Modelo',markersize=markerSize)
ax.set_xlabel('Días')  
ax.set_ylabel('USD/COP [COP]')
ax.grid(True)
ax.legend()
# ax.legend(prop={'size': 20})

twin_axes = ax.twinx() 
twin_axes.plot(D, errorAbsoluto,color='m')
twin_axes.set_ylabel('Error Absoluto [COP]' ,color='m')

plt.show()
```

```python
errorAbsoluto=y_test-y_testPrediction
print(f'Error rms:{np.sqrt(np.dot(errorAbsoluto,errorAbsoluto))}')
D=D[:len(errorAbsoluto)]
# Graficación

markerSize=3

fig,ax=plt.subplots()
ax.plot(D,y_test, 'ro', label='Real',markersize=markerSize)
ax.plot(D,y_testPrediction,'go',label='Predicción',markersize=markerSize)
ax.set_xlabel('Días')  
ax.set_ylabel('USD/COP [COP]')
ax.grid(True)
ax.legend()
# ax.legend(prop={'size': 20})

twin_axes = ax.twinx() 
twin_axes.plot(D, errorAbsoluto,color='m')
twin_axes.set_ylabel('Error Absoluto [COP]' ,color='m')

plt.show()
```

#### Análisis

Para términos prácticos el Error absoluto rms de ambos procedimiento dió lo mismo, el tiempo de entrenamiento de este último fue superior pero se obtuvo una fómula más compacta:
